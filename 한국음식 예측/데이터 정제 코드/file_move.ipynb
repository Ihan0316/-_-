```python
import os
import shutil

# 1. ê²½ë¡œ ì„¤ì •
# ----------------------------------------------------------------
root_directory = '../ë°ì´í„°/data/kfood'

# ì†Œë¶„ë¥˜ í´ë”ë“¤ì„ ì˜®ê²¨ë†“ì„ ëª©ì ì§€ í´ë”
# (ì˜ˆ: 'C:/Users/User/Desktop/ì „ì²˜ë¦¬ëœ_ì´ë¯¸ì§€')
# í´ë”ê°€ ì—†ìœ¼ë©´ ìë™ìœ¼ë¡œ ìƒì„±ë©ë‹ˆë‹¤.
destination_directory = ('../ë°ì´í„°/data/kfood_ì›ë³¸')
# ----------------------------------------------------------------


# ëª©ì ì§€ í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)
    print(f"'{destination_directory}' í´ë”ë¥¼ ìƒˆë¡œ ë§Œë“¤ì—ˆìŠµë‹ˆë‹¤.")

print(f"'{root_directory}'ì—ì„œ ìŠ¤ìº”ì„ ì‹œì‘í•©ë‹ˆë‹¤...")

# ëŒ€ë¶„ë¥˜ í´ë” ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
try:
    top_level_folders = [f for f in os.listdir(root_directory) if os.path.isdir(os.path.join(root_directory, f))]
except FileNotFoundError:
    print(f"ğŸš¨ ì˜¤ë¥˜: '{root_directory}' ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”.")
    top_level_folders = []


moved_count = 0
# ê° ëŒ€ë¶„ë¥˜ í´ë” ìˆœíšŒ
for top_folder in top_level_folders:
    top_folder_path = os.path.join(root_directory, top_folder)

    # ëŒ€ë¶„ë¥˜ í´ë” ì•ˆì˜ ì†Œë¶„ë¥˜ í´ë” ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    sub_level_folders = [f for f in os.listdir(top_folder_path) if os.path.isdir(os.path.join(top_folder_path, f))]

    # ê° ì†Œë¶„ë¥˜ í´ë”ë¥¼ ëª©ì ì§€ í´ë”ë¡œ ì´ë™
    for sub_folder in sub_level_folders:
        source_path = os.path.join(top_folder_path, sub_folder)
        destination_path = os.path.join(destination_directory, sub_folder)

        try:
            # ë§Œì•½ ëª©ì ì§€ í´ë”ì— ì´ë¯¸ ê°™ì€ ì´ë¦„ì˜ í´ë”ê°€ ìˆë‹¤ë©´, ì¶©ëŒì„ ë°©ì§€í•˜ê¸° ìœ„í•´ ì´ë¦„ ë³€ê²½
            counter = 1
            original_destination_path = destination_path
            while os.path.exists(destination_path):
                new_name = f"{sub_folder}_{counter}"
                destination_path = os.path.join(destination_directory, new_name)
                counter += 1

            if original_destination_path != destination_path:
                 print(f"   [ì£¼ì˜] ì´ë¦„ ì¶©ëŒ: '{sub_folder}' -> '{os.path.basename(destination_path)}'(ìœ¼)ë¡œ ë³€ê²½")


            # í´ë” ì´ë™!
            shutil.move(source_path, destination_path)
            print(f"âœ… '{source_path}'  ->  '{destination_path}' (ì´ë™ ì™„ë£Œ)")
            moved_count += 1

        except Exception as e:
            print(f"âŒ '{source_path}' ì´ë™ ì‹¤íŒ¨: {e}")

print("-" * 50)
print(f"ì´ {moved_count}ê°œì˜ ì†Œë¶„ë¥˜ í´ë”ë¥¼ '{destination_directory}'(ìœ¼)ë¡œ ì´ë™í–ˆìŠµë‹ˆë‹¤. âœ¨")
```
