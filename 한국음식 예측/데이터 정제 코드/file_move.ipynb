```python
import os
import shutil

# 1. 경로 설정
# ----------------------------------------------------------------
root_directory = '../데이터/data/kfood'

# 소분류 폴더들을 옮겨놓을 목적지 폴더
# (예: 'C:/Users/User/Desktop/전처리된_이미지')
# 폴더가 없으면 자동으로 생성됩니다.
destination_directory = ('../데이터/data/kfood_원본')
# ----------------------------------------------------------------


# 목적지 폴더가 없으면 생성
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)
    print(f"'{destination_directory}' 폴더를 새로 만들었습니다.")

print(f"'{root_directory}'에서 스캔을 시작합니다...")

# 대분류 폴더 목록 가져오기
try:
    top_level_folders = [f for f in os.listdir(root_directory) if os.path.isdir(os.path.join(root_directory, f))]
except FileNotFoundError:
    print(f"🚨 오류: '{root_directory}' 경로를 찾을 수 없습니다. 경로를 확인해주세요.")
    top_level_folders = []


moved_count = 0
# 각 대분류 폴더 순회
for top_folder in top_level_folders:
    top_folder_path = os.path.join(root_directory, top_folder)

    # 대분류 폴더 안의 소분류 폴더 목록 가져오기
    sub_level_folders = [f for f in os.listdir(top_folder_path) if os.path.isdir(os.path.join(top_folder_path, f))]

    # 각 소분류 폴더를 목적지 폴더로 이동
    for sub_folder in sub_level_folders:
        source_path = os.path.join(top_folder_path, sub_folder)
        destination_path = os.path.join(destination_directory, sub_folder)

        try:
            # 만약 목적지 폴더에 이미 같은 이름의 폴더가 있다면, 충돌을 방지하기 위해 이름 변경
            counter = 1
            original_destination_path = destination_path
            while os.path.exists(destination_path):
                new_name = f"{sub_folder}_{counter}"
                destination_path = os.path.join(destination_directory, new_name)
                counter += 1

            if original_destination_path != destination_path:
                 print(f"   [주의] 이름 충돌: '{sub_folder}' -> '{os.path.basename(destination_path)}'(으)로 변경")


            # 폴더 이동!
            shutil.move(source_path, destination_path)
            print(f"✅ '{source_path}'  ->  '{destination_path}' (이동 완료)")
            moved_count += 1

        except Exception as e:
            print(f"❌ '{source_path}' 이동 실패: {e}")

print("-" * 50)
print(f"총 {moved_count}개의 소분류 폴더를 '{destination_directory}'(으)로 이동했습니다. ✨")
```
